```python
cancer2015 = df.query(' Year == 2015 ')
Stomachcancer2015 = cancer2015.sort_values(by = 'Stomach cancer', ascending=False).head(10)
plt.figure(figsize=(16,8))
sns.barplot(x=Stomachcancer2015['Stomach cancer'],y=Stomachcancer2015['Entity'],data=Stomachcancer2015,palette='RdGy')
plt.title('Top countries with highest number of Stomach cancer death in 2015',fontsize=18, color = 'Blue')
plt.ylabel('Country',fontsize=15)
plt.xlabel('Stomach cancer',fontsize=15)
```

```python
Breastcancer_Top = df.groupby('Entity').sum()['Breast cancer'].nlargest(10)
Breastcancer_Top = pd.DataFrame(Breastcancer_Top)
Stomachcancer_Top = df.groupby('Entity').sum()['Stomach cancer'].nlargest(10)
Stomachcancer_Top = pd.DataFrame(Stomachcancer_Top)
```

```python
fig = plt.figure(figsize=(27,18))

sub1 = fig.add_subplot(2,2,1)
sub1.set_xlabel('Stomach cancer')
sub1.set_title('Top countries with highest number of Stomach cancer death')
sns.barplot(x=Stomachcancer_Top['Stomach cancer'],y=Stomachcancer_Top.index,data=Stomachcancer_Top,palette='viridis')

sub1 = fig.add_subplot(2,2,2)
sub1.set_xlabel('Year')
sub1.set_title('Stomach cancer cancer by Year')
df.groupby('Year').sum()['Stomach cancer'].plot(linewidth=4, color = 'Blue')
plt.legend(bbox_to_anchor=(1, 1), loc=1)

sub1 = fig.add_subplot(2,2,3)
sub1.set_xlabel('Year')
sub1.set_title('Top countries with highest number of Breast cancer death')
sns.barplot(x=Breastcancer_Top['Breast cancer'],y=Breastcancer_Top.index,data=Breastcancer_Top,palette='plasma')

sub1 = fig.add_subplot(2,2,4)
sub1.set_xlabel('None')
sub1.set_title('Breast cancer by Year')
df.groupby('Year').sum()['Breast cancer'].plot(linewidth=4,color='Green')
plt.legend(bbox_to_anchor=(1, 1), loc=1)
```

```python
India = df.query(' Entity == "India" ')[['Entity', 'Year','Breast cancer','Prostate cancer','Stomach cancer','Liver cancer']]
fig = plt.figure(figsize=(27,18))

sub1 = fig.add_subplot(2,2,1)
sub1.set_xlabel('Year')
sub1.set_title('Prostate cancer death by Year in India')
India.groupby('Year').sum()['Prostate cancer'].plot(linewidth=4,color='blue')
plt.legend(bbox_to_anchor=(1, 1), loc=1)

sub1 = fig.add_subplot(2,2,2)
sub1.set_xlabel('Year')
sub1.set_title('Breast cancer death by Year in India')
India.groupby('Year').sum()['Breast cancer'].plot(linewidth=4, color='red')
plt.legend(bbox_to_anchor=(1, 1), loc=1)

sub1 = fig.add_subplot(2,2,3)
sub1.set_xlabel('Year')
sub1.set_title('Stomach cancer death by Year in India')
India.groupby('Year').sum()['Stomach cancer'].plot(linewidth=4, color = 'orange')
plt.legend(bbox_to_anchor=(1, 1), loc=1)

sub1 = fig.add_subplot(2,2,4)
sub1.set_xlabel('Year')
sub1.set_title('Liver cancer death by Year in India')
India.groupby('Year').sum()['Liver cancer'].plot(linewidth=4, color = 'green')
plt.legend(bbox_to_anchor=(1, 1), loc=1)
```
